```python
import json
import pandas as pd
```

```python
file = open('./data/dataSources/champions_14_5_1.json')
data = json.load(file)
```

```python
stat_keys = list(data['data']['Aatrox']['stats'].keys())
columns = ['name','patch','attack','defense','magic','difficulty','tag_1','tag_2']
columns.extend(stat_keys)
champ_df = pd.DataFrame(columns=columns)
```

```python
champ_df.columns
```

```python
for champ in data['data']:
    champ_data = data['data'][champ]
    new_champ = dict()
    new_champ['name'] = champ_data['name']
    new_champ['patch'] = champ_data['version']
    new_champ['attack'] = champ_data['info']['attack']
    new_champ['defense'] = champ_data['info']['defense']
    new_champ['magic'] = champ_data['info']['magic']
    new_champ['difficulty'] = champ_data['info']['difficulty']

    new_champ['tag_1'] = champ_data['tags'][0]
    if len(champ_data['tags']) > 1:
        new_champ['tag_2'] = champ_data['tags'][1]
    else:
        new_champ['tag_2'] = ''

    for key,value in list(champ_data['stats'].items()):
        new_champ[key] = value


    champ_df = pd.concat([champ_df, pd.DataFrame([new_champ])], ignore_index=True)
```

```python
champ_df.to_csv('./data/gameData/p_14_5_1_champs.csv')
```
