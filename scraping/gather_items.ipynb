```python
import json
import pandas as pd
```

```python
patch = '1451'
file = open(F'./data/dataSources{patch}/items.json')
data = json.load(file)
```

```python
columns = ['name','lol_number','into','from','cost','purchasable','sell','consumed','in_store','consumeOnFull','requiredChampion','requiredAlly']
item_df = pd.DataFrame(columns=columns)
item_df.columns
```

```python
for item in data['data']:
    item_data = data['data'][item]
    
    if item_data['name'] in item_df['name'].unique():
        continue

    new_item = dict()
    new_item['name'] = item_data['name']
    new_item['lol_number'] = item
    
    if 'into' in item_data:
        new_item['into'] = item_data['into']
    else:
        new_item['into'] = ''

    if 'from' in item_data:
        new_item['from'] = item_data['from']
    else:
        new_item['from'] = ''

    new_item['cost'] = item_data['gold']['total']
    new_item['purchasable'] = item_data['gold']['purchasable']
    new_item['sell'] = item_data['gold']['sell']

    if 'consumed' in item_data:
        new_item['consumed'] = item_data['consumed']
    else:
        new_item['consumed'] = False

    if 'inStore' in item_data:
        new_item['in_store'] = item_data['inStore']
    else:
        new_item['in_store'] = True

    if 'consumeOnFull' in item_data:
        new_item['consumeOnFull'] = item_data['consumeOnFull']
    else:
        new_item['consumeOnFull'] = False
    
    if 'requiredChampion' in item_data:
        new_item['requiredChampion'] = item_data['requiredChampion']
    else:
        new_item['requiredChampion'] = ''

    if 'requiredAlly' in item_data:
        new_item['requiredAlly'] = item_data['requiredAlly']
    else:
        new_item['requiredAlly'] = ''


    item_df = pd.concat([item_df, pd.DataFrame([new_item])], ignore_index=True)
```
