```python
import pandas as pd
from pycaret import classification
```

```python
df = pd.read_csv('../../data/feature_sets/v0.csv')
```

```python
classification_setup = classification.setup(data= df, target='red_team_outcome',use_gpu=True,feature_selection = True,remove_multicollinearity = True)
```

```python
best = classification.compare_models()
```

```python
best
```

```python
classification.plot_model(best, plot = 'feature')
```

```python
classification.interpret_model(best, plot = 'summary')
```

```python
classification.save_model(best, 'saved_dt_model')
```
