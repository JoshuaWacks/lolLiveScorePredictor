```python
league_df = pd.DataFrame(columns=['id','slug','name'])
```

```python
temp_list = []
for league in json_data['data']['leagues']:
    new_row = {'id': league['id'],'slug': league['slug'],'name': league['name']}
    temp_list.append(new_row)

league_df = pd.DataFrame.from_dict(temp_list)
```
